```python
import numpy as np
import pandas as pd
```

```python
df_can = pd.read_excel('https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true',
                      sheet_name = 'Canada by Citizenship',
                      skiprows=range(20),
                      skipfooter=2)
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.sum(axis=1)
```

```python
df_can.head(2)
```

```python
df_can.columns
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
```

```python
df_can.columns
```

```python
df_can.head(2)
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.head(3)
```

```python
df_can.loc[:, :]
```

```python
df_can.loc[:, :].sum()
```

```python
df_can.append(df_can.loc[:, :].sum(), ignore_index=True)
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
import matplotlib as mpl 
```

```python
import matplotlib.pyplot as plt
```

```python
japan = df_can.loc['Japan', years]
japan
```

```python
japan.plot()
```

```python
japan.index = japan.index.map(int)
japan.plot(kind='line')
plt.title('perkembangan imigran jepang di kanada')
plt.xlabel('tahun')
plt.ylabel('populasi')
plt.text(2000, 800, 'doraemon')
plt.show()
```

```python
japan_singapore = df_can.loc[['Japan', 'Singapore'], years]
japan_singapore
```

```python
japan_singapore.plot()
```

```python
df_js = japan_singapore.transpose()
df_js
```

```python
df_js.plot()
```

```python
df_js.plot(kind='line')
plt.title('perbandingan imigran jepang dan singapura')
plt.xlabel('tahun')
plt.ylabel('populasi')
plt.show()
```
