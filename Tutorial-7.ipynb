```python
orang_terkaya = {
    "budi hartono": 20.5,
    "michael hartono": 19.7,
    "sri prakash lohia": 6.5,
    "prajogo pangestu": 6.5,
    "chairul tanjung": 4.8,
    "tahir": 3.3,
    "eddy sariaatmadja": 3,
    "jerry ng": 2.5,
    "martua sitorus": 2,
    "theodore rahmat": 1.7
}
orang_terkaya
```

```python
nama_orang_terkaya = orang_terkaya.keys()
nama_orang_terkaya
```

```python
y_pos = np.arange(len(nama_orang_terkaya))
kekayaan = orang_terkaya.values()
y_pos
```

```python
plt.figure(figsize=(18, 5))
plt.bar(y_pos, kekayaan, align='center', alpha=0.5)
plt.xticks(y_pos, nama_orang_terkaya)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(y_pos, kekayaan, align='center', alpha=0.5)
plt.xticks(y_pos, nama_orang_terkaya, rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.barh(y_pos, kekayaan, align='center', alpha=0.5)
plt.yticks(y_pos, nama_orang_terkaya)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('A single plot')
```

```python
fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y)
axs[1].plot(x, -y)
```

```python
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
ax3.plot(x, 2*y)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
fig.suptitle('Vertically stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
ax3.plot(x, 2*y)
fig.set_size_inches((20,4))
```
