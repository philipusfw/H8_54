```python
df2 = df2.dropna(subset=['Height_cm'])
```

```python
def convert_feet_to_cm(x):
    if "." in x: # 182.8
        return int(float(x))
    if "'" not in x and ";" not in x: # 182
        return int(x)
    # 5'1
    if "'" in x:
        feet_inch = x.split("'")
    elif ";" in x:
        feet_inch = x.split(";")
    feet = int(feet_inch[0])
    inch = int(feet_inch[1])
    height = 0
    height += feet * 30.48
    height += inch * 2.54
    return int(height)
```

```python
df2
```

```python
df2['Height_cm2'] = df2['Height_cm'].apply(convert_feet_to_cm)
df2['Height_cm2']
```

```python
height_mean = df2['Height_cm2'].mean()
height_mean
```
