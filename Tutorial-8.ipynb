```python
import numpy as np
import pandas as pd
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('alice_novel.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
alice_wc = WordCloud(max_words=2000, background_color='white', stopwords = stopwords)
```

```python
alice_wc.generate(alice_novel)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
stopwords.add('said')
alice_wc.generate(alice_novel)
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```
